```python
%reset -f
import os

from ROOT import *
from Roottols.Helper import *
from Roottols.RootHelper import FillGraph, FillGraphAsymmErrrors, MkBranch
from Roottols.Plotter import *
from Roottols.Plotter import Tplot as pl, tplotconf as pc
import Roottols.TreePlotHelper as ph
from Roottols.Functions import *
from Roottols.PyRootPlotManager import PyRootPlotManager

%matplotlib widget
%jsroot on
```

```python
infile = ROOT.TFile('../data/prova.root')
intree = infile.Get('T')

cc = Canva()
cc.loadstyle()

defcf = pc(dict(tree=intree, cman=cc))
```

```python
dumax = 32
dommax = 20
rngq = (2000,0,200)
rngqsmall = (120,0,40)
rngdom = (20,0,20)
rangepmt = (32,0,32)
rngdt = (1000,-1000,1000)
rngqsum = (1000,0,1000)
rngtdiff = (200,-50,50) 
rngy = (500,0,500)
rngx = (600,-300,300)
rngr = (600,0,600)
rngz = (700,0,700)
rngd = (1200,0,600)
```

```python
_=pl([defcf,dict(
        val=        'D2',
        rng=        rngd,
        cut=        '',
        opt=        'zcol text1',
        axlab=      'D[m]',
        csiz=       (800,500),
        rebin =     1,
        optstat=    10,
        mcolor=     ROOT.kBlue,        
        msize=      1,
        mstyle=     20,
)])
```

```python
_=pl([defcf,dict(
        val=        ['Y:X','Z:sqrt(X*X+Y*Y)'],
        rng=        [(*rngx,*rngy),(*rngr,*rngz)],
        cut=        '',
        opt=        'zcol text1',
        axlab=      [('X [mm]; Y [mm]'), ('R [mm]; Z [mm]')],
        csiz=       (1000,1200),
        cdiv=       (1,2),
        rebin =     (4,4),
        optstat=    10,
        mcolor=     ROOT.kBlack,        
        msize=      [2,0],
)])
```

```python
_=pl([defcf,dict(
        val=        'dT:D2',
        rng=        (*rngd,*rngdt),
        cut=        '',
        opt=        'zcol',
        axlab=      'D[m];DT[ns]',
        csiz=       (1000,500),
        rebin =     (2,2),
        optstat=    10,
)])

_=pl([defcf,dict(
        val=        'dT:D2',
        rng=        (*rngd,*rngdt),
        cut=        '',
        opt=        'prof',
        axlab=      'D[m];DT[ns]',
        csiz=       (1000,500),
        rebin =     4,
        optstat=    10,
        proc=       lambda xx, xxx: xx.Fit('pol1', '', '', 50,300)
)])
```

```python
_=pl([defcf,dict(
        val=        'npe:D2',
        rng=        (*rngd,*rngq),
        cut=        '',
        opt=        'zcol',
        axlab=      'D[m];Q[npe]',
        csiz=       (1000,500),
        rebin =     (2,2),
        optstat=    10,
)])

_=pl([defcf,dict(
        val=        'npe:D2',
        rng=        (*rngd,*rngq),
        cut=        '',
        opt=        'prof',
        axlab=      'D[m];Q[npe]',
        csiz=       (1000,500),
        rebin =     2,
        optstat=    10,
        logy=       1,
        proc=       lambda xx, xxx: xx.Fit('expo', '', '', 0, 180)
)])
```
